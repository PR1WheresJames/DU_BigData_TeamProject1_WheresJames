```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
```

```python
file_crime = 'crime.csv'
crime_df = pd.read_csv(file_crime,nrows = 100)
# crime_df.head()
# crime_df.columns ['INCIDENT_ID', 'OFFENSE_ID', 'OFFENSE_CODE', 'OFFENSE_CODE_EXTENSION',
#        'OFFENSE_TYPE_ID', 'OFFENSE_CATEGORY_ID', 'FIRST_OCCURRENCE_DATE',
#        'LAST_OCCURRENCE_DATE', 'REPORTED_DATE', 'INCIDENT_ADDRESS', 'GEO_X',
#        'GEO_Y', 'GEO_LON', 'GEO_LAT', 'DISTRICT_ID', 'PRECINCT_ID',
#        'NEIGHBORHOOD_ID', 'IS_CRIME', 'IS_TRAFFIC'],

#convert string timestamp to_datetime() 
crime_df['test_date'] = pd.to_datetime(crime_df["REPORTED_DATE"])

#call datetime functions on datetime field, year,month,qtr....etc 
crime_df['month']  = crime_df['test_date'].dt.month 
crime_df['month'] = crime_df['month'].map('{:02d}'.format)
crime_df['year']  = crime_df['test_date'].dt.year
crime_df['year-month'] = crime_df["year"].map(str) + '-' + crime_df["month"].map(str)
crime_df.head()
```
